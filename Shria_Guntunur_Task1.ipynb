```python
#importing necessary libraries
import pandas as pd
import numpy as np  
import matplotlib.pyplot as plt
```

```python
#importing data
url = "http://bit.ly/w-data"
data = pd.read_csv(url)
print("Data imported successfully!")
```

```python
#displaying data
data.head(25)
```

```python
#plotting data
data.plot(x='Hours', y='Scores',style='o')  
plt.title('Hours vs Percentage')  
plt.xlabel('Hours Studied')  
plt.ylabel('Percentage Score')  
plt.show()
```

```python
X = data.iloc[:, :-1].values #get a copy of dataset exclude last column, -1=1st col, first : is for column
Y = data.iloc[:, 1].values #get array of dataset in column 1st
```

```python
#divide the dataset into training and test set
from sklearn.model_selection import train_test_split 
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=1/4, random_state=0)#25 percent test size
```

```python
from sklearn.linear_model import LinearRegression
regressor = LinearRegression()
regressor.fit(X_train, Y_train)
```

```python
# Visualizing the Training set results
train_v = plt
train_v.scatter(X_train, Y_train, color='red')
train_v.plot(X_train, regressor.predict(X_train), color='blue')
train_v.title('Hours VS Percentage (Training set)')
train_v.xlabel('Hours studied')
train_v.ylabel('Percentage scored')
train_v.show()

# Visualizing the Test set results
test_v = plt
test_v.scatter(X_test, Y_test, color='red')
test_v.plot(X_train, regressor.predict(X_train), color='blue')
test_v.title('Hours VS Percentage (Training set)')
test_v.xlabel('Hours studied')
test_v.ylabel('Percentage scored')
test_v.show()
```

```python
#predict percentage scored if number of hours studied is 9.25
hours = [[9.25]]
pred = regressor.predict(hours)
pred
```

```python
#predict for the entire test set
y_pred = regressor.predict(X_test)
y_pred
```

```python
#display the actual and predicted values 
df = pd.DataFrame({'Actual': Y_test, 'Predicted': y_pred})  
df 
```

```python
#calculate the mean absolute error
from sklearn import metrics  
print('Mean Absolute Error:', 
      metrics.mean_absolute_error(Y_test, y_pred)) 
```
